```python
data=[
[1,2,3,'Good'],
[4,np.nan,6,'Good'],
[7,np.nan,9,'Excellent'],
[10,11,12,],
[13,14,15,'Excellent'],
[16,17,np.nan,'Fair'],
[19,12,12,'Excellent'],
[20,11,23,'Fair']]
df3=pd.DataFrame(data,columns=["A",'B','C','D'])
df3
```

```python
#replacing with median
imputer=SimpleImputer(strategy="median",missing_values=np.nan)
imputer=imputer.fit(df3[["B","C"]])
df3[["B","C"]]=imputer.transform(df3[["B","C"]])
df3
```
